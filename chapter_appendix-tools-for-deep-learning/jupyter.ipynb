```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)
df.head()
```

```python
df.isnull().sum()
df.duplicated().sum()
```

```python
df_filled = df.fillna(df.mean(numeric_only=True))
df_filled.head()
```

```python
df_clean = df_filled.drop_duplicates()
df_clean.head()
```

```python
import matplotlib.pyplot as plt
df_clean.boxplot(column='Marks')
```

```python
df_clean = df_clean[df_clean['Marks'] < 100]
```

```python
df_clean['Performance'] = df_clean['Marks'] * df_clean['Attendance']
df_clean.head()
```
